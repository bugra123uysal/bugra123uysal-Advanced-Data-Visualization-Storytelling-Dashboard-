```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
Ay bazlı satış verisi
- Ürün bazlı satış oranları
- Bölge bazlı performans
 '''

ay=["ocak","şubat","mart","nisan"]
satis1 = [100,200,150,300]
satis2 = [90,250,120,280]
satis3 = [50,180,220,260]
satis4 = [70,160,140,310]



ürün=["sigara","viski","et","kola","bilgisayar"]
ürün_b_satış=[1000,2000,1500,900,700]

ürün_bazlı_satış=pd.DataFrame({"ürün":ürün,"übso":ürün_b_satış})

bölge=["karadeniz","ege","iç anadolu","marmara","akdeniz"]
bolge_b_satış=[4000,2000,6000,7000,1000]

bolge_bazlı_satış=pd.DataFrame({"bölge":bölge,"bbp":bolge_b_satış})
```

```python
'''
 `plt.subplots()` ile **2x2 grid**
- Ortak başlık (`suptitle`)
- Ortak eksen kullanımı
 '''
fig,axes=plt.subplots(2,2,figsize=(10,10), sharex=True)
fig.suptitle("Ay Bazlı Satış")
axes[0,0].bar(ay,satis1)
axes[0,0].set_title("satış-1")

axes[0,1].bar(ay,satis2)
axes[0,1].set_title("satış-2")

axes[1,0].bar(ay,satis3)
axes[1,0].set_title("Satış-3")


axes[1,1].bar(ay,satis4)
axes[1,1].set_title("Satış-4")


plt.tight_layout()
plt.show()

```

```python
from matplotlib import text
from matplotlib.text import Annotation
'''
- Grid ayarları
- Annotation (önemli noktaları işaretleme)
- Text ekleme

 '''
plt.bar(ürün_bazlı_satış["ürün"], ürün_bazlı_satış["übso"])
plt.xlabel("Ürünler")
plt.ylabel("Satış Miktarı")
plt.grid(True)

plt.annotate(
    "En yüksek satış",
    xy=("viski", 2000),
    xytext=("et", 2200),
    arrowprops=dict(facecolor="red")


)
plt.annotate(
    "en düşük satış",
    xy=("bilgisayar", 700),
    xytext=("kola", 1100),
    arrowprops=dict(facecolor="yellow")

)

plt.show()
```

```python
'''
Renk paleti tutarlılığı
- Font boyutu hiyerarşisi
- Arka plan düzeni
 '''
plt.figure(figsize=(8, 6))
plt.bar(bolge_bazlı_satış["bölge"], bolge_bazlı_satış["bbp"], color="steelblue")
plt.xlabel("Bölge", fontsize="12")
plt.ylabel("Satış Miktarı",fontsize="12")
plt.title("Bölge Bazlı Satış Miktarı", fontsize="16")
plt.grid(True, linestyle="--",alpha=0.5)
plt.xticks(fontsize="9")
plt.yticks(fontsize="9")
plt.legend()
plt.show()
```

```python
'''
Line plot
- Bar plot
- Scatter plot
- Pie veya Histogram
 '''
fig,axes=plt.subplots(2,
                      2,
                      figsize=(14,10))
# bar
axes[0,0].bar(ürün_bazlı_satış["ürün"], ürün_bazlı_satış["übso"],color="steelblue")
axes[0,0].set_xlabel("Ürünler")
axes[0,0].set_ylabel("Satış Miktarı")
axes[0,0].grid(True)
axes[0,0].set_title("Bar Plot",fontsize=14)

# line


axes[0,1].scatter(ürün_bazlı_satış["ürün"], ürün_bazlı_satış["übso"], color="red")
axes[0,1].set_xlabel("Ürünler")
axes[0,1].set_ylabel("Satış Miktarı")
axes[0,1].grid(True)
axes[0,1].set_title("Scatter Plot",fontsize=14)



# pie pasta
axes[1,0].pie(ürün_bazlı_satış["übso"], labels=ürün_bazlı_satış["ürün"],autopct="%1.1f%%" )
axes[1,0].set_title("Pie Chart",fontsize=14)


# scatter plot
axes[1,1].plot(ürün_bazlı_satış["ürün"], ürün_bazlı_satış["übso"], color="red")
axes[1,1].set_xlabel("Ürünler")
axes[1,1].set_ylabel("Satış Miktarı")
axes[1,1].grid(True)
axes[1,1].set_title("Scatter Plot",fontsize=14)
marker="*"

plt.suptitle("Ürün Bazlı Satış Miktarı dasbordu",fontsize=17)

plt.tight_layout()
plt.show()


```
